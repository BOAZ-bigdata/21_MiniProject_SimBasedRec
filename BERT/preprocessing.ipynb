```python
from datasets import load_dataset

data_st_plus = load_dataset("lbox/lbox_open", "statute_classification_plus")
```

```python
import pandas as pd

columns_to_drop = ['id', 'casetype', 'casename']

train_df = pd.DataFrame(data_st_plus['train'])
train_df = train_df.drop(columns=columns_to_drop, axis=1)

validation_df = pd.DataFrame(data_st_plus['validation'])
validation_df = validation_df.drop(columns=columns_to_drop, axis=1)

test_df = pd.DataFrame(data_st_plus['test'])
test_df = test_df.drop(columns=columns_to_drop, axis=1)
```

```python
import pandas as pd

result_df = pd.concat([train_df, validation_df, test_df], ignore_index=True)
```

```python
def list_to_string(lst):
    return ', '.join(map(str, lst))

result_df['statutes'] = result_df['statutes'].apply(list_to_string)
```

```python
def split_and_create_columns(row):
    words = row['statutes'].split(',')
    for i, word in enumerate(words):
        row[f'word_{i+1}'] = word
    return row

result_df = result_df.apply(split_and_create_columns, axis=1)
result_df
```

```python
target_word = '법'

def cut_until_target(text, target):
    if type(text) == str:
      index = text.rfind(target)
      if index != -1:
          return text[:index + len(target)]
      else:
          return text

result_df['word_1'] = result_df['word_1'].apply(lambda x: cut_until_target(x, target_word))
```

```python
import pandas as pd

value_counts = result_df['word_1'].value_counts()

values_to_remove = value_counts[value_counts < 10].index
result_df = result_df[~result_df['word_1'].isin(values_to_remove)]

result_df
```

```python
result_df = result_df[['word_1', 'facts']]
```
